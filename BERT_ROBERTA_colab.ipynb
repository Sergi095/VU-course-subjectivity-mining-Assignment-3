```python
from simpletransformers.classification import ClassificationModel
```

```python
olid_train_small = pd.read_csv('/content/drive/MyDrive/Datasets assingmet 3 hate speech/olid-train-small.csv', sep=',')
hasoc_train = pd.read_csv('/content/drive/MyDrive/Datasets assingmet 3 hate speech/hasoc-train.csv', sep=',')
olid_test  = pd.read_csv('/content/drive/MyDrive/Datasets assingmet 3 hate speech/olid-test.csv', sep=',')
```

```python
def precision(tp, fp):
    prec = tp/(tp+fp)
    return prec

def recall(tp, fn):
    rec = tp/(tp+fn)
    return rec

def f1(p, r):
    f = 2* ((p*r)/(p+r))
    return f


def macro_av(score1, score2):
    macro = np.mean([score1, score2])
    return macro

def make_one_list(tp, fp, tn, fn): 
    '''
    RETURNS a dictionary with scores
    '''
    prec_o = precision(tp,fp)
    rec_o = recall(tp,fn)
    f_o = f1(prec_o, rec_o)
    prec_n = precision(tn, fn)
    rec_n = recall(tn, fp)
    f_n = f1(prec_n, rec_n)
    macro_p = macro_av(prec_o, prec_n)
    macro_r = macro_av(rec_o, rec_n)
    macro_f = macro_av(f_o, f_n)
    results_dict = {"prec off": prec_o, "recall off": rec_o, "f1 o": f_o,
                    "prec nof": prec_n, "recall nof": rec_n, "f1_n": f_n,
                    "mac p": macro_p, "mac r": macro_r, "mac_f1": macro_f}
    return results_dict
```

```python
model_bert = ClassificationModel('bert', 'bert-base-cased',use_cuda = True) 
```

```python
model_roberta = ClassificationModel("roberta", "roberta-base",use_cuda = True)
```

```python
# changing column names !
hasoc_train.columns = ['id','text','labels']
olid_train_small.columns = ['id','text','labels']
olid_test.columns = ['id','text','labels']
```

```python
model_bert.train_model(hasoc_train, num_train_epochs = 5, output_dir = "/content/outputs_bert_cross")
```

```python
model_roberta.train_model(hasoc_train, num_train_epochs = 5, output_dir = "/content/outputs_roberta_cross")
```

```python
result, model_outputs, wrong_predictions = model_bert.eval_model(olid_test)
```

```python
print('BERT CROSS DOMAIN')
print(make_one_list(result['tp'], result['fp'], result['tn'], result['fn']))
```

```python
result, model_outputs, wrong_predictions = model_roberta.eval_model(olid_test)
```

```python
print('ROBERTA CROSS DOMAIN')
print(make_one_list(result['tp'], result['fp'], result['tn'], result['fn']))
```
